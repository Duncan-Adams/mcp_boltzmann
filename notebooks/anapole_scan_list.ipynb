## Here is a short script to generate text lists to control the scan parameters

```python
import os
import sys

if '../' not in sys.path:
    sys.path.append('../')
```

```python
import numpy as np
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]
```

### Scan range for m_chi

```python
mchi_min_MeV = 1e-3
mchi_max_MeV = 1e3

num_masses = 50

masses = np.geomspace(mchi_min_MeV, mchi_max_MeV, num_masses)
```

### Define R

```python
R = 3
```

### Scan range for Q

```python
Q_min = 1e-11
Q_max = 1e-4
num_charges = 50

charges = np.geomspace(Q_min, Q_max, num_charges)
```

```python
scan = np.zeros((len(masses)*len(charges), 3))
index = 0

for m_chi in masses:
    for Q in charges:
        scan[index, 0] = m_chi
        scan[index, 1] = R
        scan[index, 2] = Q 
        index += 1
```

### annhilation scan file

```python
with open(f'../input/anapole_mcp_annihilation_R_{R}.csv', 'w') as f:
    for mass in masses:
        print(f'{(R*mass):.3e}, {1e-2*R*mass:.3e}, {1e5*R*mass:.3e}', file=f)
    
```

```python
anapole_scan_file = 'anapole_Neff_scan.txt'
anapole_path = os.path.join('../input/', anapole_scan_file )

with open(anapole_path, 'w') as out_file:
    np.savetxt(out_file, scan, fmt=('%.5e', '%d', '%.5e'), delimiter=', ', header='m[MeV] , R, Q')
```
