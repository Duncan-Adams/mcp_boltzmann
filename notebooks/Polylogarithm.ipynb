# Make plots comparing our approximate polylog's with the mpmath one

```python
x_plot = np.linspace(-0.999, 0.999)
n_plot = np.array([0, 1, 2, 3])
```

```python
plf = plog.polylog_fast(n_plot, x_plot)
```

```python
plt.plot(x_plot, plf[0], label='polylog_fast')
plt.plot(x_plot, plog.polylog_vec(0, x_plot), label='mpmath', linestyle='dashed')
plt.xlim(-1, 0)
plt.ylim(-0.5, 0)
plt.legend()
plt.show()

plt.plot(x_plot, plf[0], label='polylog_fast')
plt.plot(x_plot, plog.polylog_vec(0, x_plot), label='mpmath', linestyle='dashed')
plt.xlim(0, 1)
plt.yscale('log')

plt.legend()
plt.show()
```
